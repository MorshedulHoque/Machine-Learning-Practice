Now calculate the mean absulate error

```python
from sklearn.metrics import mean_absolute_error

val_mae = mean_absolute_error(val_Y, val_prediction)
```

```python
val_mae
```

```python
def get_mae(max_leaf_nodes, train_X, val_X, tarin_Y, val_Y):
  model3 = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
  model3.fit(train_X, train_Y)
  preds_val = model3.predict(val_X)
  mae = mean_absolute_error(val_Y, preds_val)
  return mae
```

```python
import numpy as np
```

```python
candidate_max_leaf_nodes = [5, 25, 50, 100, 250, 500]

best_mae = np.Inf
for candidate in candidate_max_leaf_nodes:
  mae = get_mae(candidate, train_X, val_X, train_Y, val_Y)
  if mae < best_mae:
    best_mae = mae
    best_tree_size = candidate
```

```python
import numpy as np
import matplotlib.pyplot as plt

candidate_max_leaf_nodes = [5, 25, 50, 100, 250, 500]

best_mae = np.Inf
mae_values = []  # Store MAE values for each candidate
best_tree_size = None

for candidate in candidate_max_leaf_nodes:
    mae = get_mae(candidate, train_X, val_X, train_Y, val_Y)
    mae_values.append(mae)  # Store MAE for this candidate
    if mae < best_mae:
        best_mae = mae
        best_tree_size = candidate

# Set custom height and width for the plot canvas
plt.figure(figsize=(10, 6))  # Adjust the values for your desired dimensions

# Create a plot to visualize the relationship between tree size and MAE
plt.plot(candidate_max_leaf_nodes, mae_values, marker='o', linestyle='-')
plt.xlabel('Max Leaf Nodes')
plt.ylabel('Mean Absolute Error (MAE)')
plt.title('MAE vs. Max Leaf Nodes')
plt.grid(True)

# Highlight the best tree size with a smaller arrow
plt.annotate(f'Best MAE ({best_tree_size}): {best_mae:.2f}', xy=(best_tree_size, best_mae),
             xytext=(best_tree_size + 50, best_mae + 1),
             arrowprops=dict(arrowstyle='->', color='red', shrinkA=0.05, shrinkB=0.05, lw=1))

plt.show()
```
